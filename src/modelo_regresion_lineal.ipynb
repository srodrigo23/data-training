```python
import pandas as pd
datos = pd.read_csv('ingreso.csv')
datos
```

```python
import matplotlib.pyplot as plt
# grafica de dispersion
plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanlaes trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="pink")
plt.show()
```

## Creación del Modelo de Regresion Lineal Simple

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()
horas = datos["horas"].values.reshape((-1, 1))#vectorzando, vectores de un solo elemento.
modelo = regresion.fit(horas, datos['ingreso'])#ajustar una linea, horas es una var. independiente, datos['ingreso'] variable dependiente
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_) #por cada hora trabajada el sueldo aumenta en 3mil pesos

entrada = [[39], [40], [43], [44]]
prediction = modelo.predict(entrada)

plt.scatter(entrada, prediction, color="red")
plt.plot(entrada, prediction, color="black")

plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="pink")
plt.show()
```
