```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Ajustando un modelo lineal

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

Encontramos una variable discreta

```python
sns.regplot(x="size", y="tip", data=tips, x_estimator=np.mean)
#grafico de medias las patitas son el hancho de media para cada estimacion
```
